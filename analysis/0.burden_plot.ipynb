- Import the libraries you need (pandas, matplotlib, seaborn)

```python
import pandas
import matplotlib
import seaborn

from pathlib import Path
```

```python
seaborn.set_style('whitegrid')
seaborn.set_palette('colorblind')
```

```python
mtb_df = pandas.read_csv('../data/ihme_microbe/0.ihme_mtb_data.csv')
spneumoniae_df = pandas.read_csv('../data/ihme_microbe/1.ihme_s.pneumoniae_data.csv')
saureus_df = pandas.read_csv('../data/ihme_microbe/2.ihme_s.aureus_data.csv')
kpneumoniae_df = pandas.read_csv('../data/ihme_microbe/3.ihme_k.pneumoniae_data.csv')
ecoli_df = pandas.read_csv('../data/ihme_microbe/4.ihme_e.coli_data.csv')
ihme_microbe_df = [mtb_df, spneumoniae_df, saureus_df, kpneumoniae_df, ecoli_df]
all_ihme_microbe_df = pandas.concat(ihme_microbe_df)
```

```python
seaborn.catplot(data=all_ihme_microbe_df, kind="bar", x="Location", y="Value", hue="Pathogen")
```

```python
subset = all_ihme_microbe_df[all_ihme_microbe_df['Location'].isin(['Canada', 'United States of America', 'France', 
                                                                   'China', 'Russian Federation'])]
```

```python
seaborn.catplot(data=subset, kind="bar", y="Location", x="Value", hue="Pathogen")
```
